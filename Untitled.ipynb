```python
import os
import cv2
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator

def augment_images(image_folder, output_folder, num_augmented_images, target_size=(224, 224)):
    datagen = ImageDataGenerator(
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
    )

    images = []
    for filename in os.listdir(image_folder):
        img_path = os.path.join(image_folder, filename)
        img = cv2.imread(img_path)
        if img is not None:
            img = cv2.resize(img, target_size)  # Resize the image
            images.append(img)

    images = np.array(images)
    total_augmented_images = len(images) * num_augmented_images
    print(f"Augmenting to {total_augmented_images} images")

    i = 0
    for image in images:
        image = np.expand_dims(image, 0)
        for batch in datagen.flow(image, batch_size=1, save_to_dir=output_folder, save_prefix='aug', save_format='jpeg'):
            i += 1
            if i >= total_augmented_images:
                break
        if i >= total_augmented_images:
            break

# Paths to your datasets
parkinsons_folder = 'C:/Users/Toshal/Desktop/Project/parkinsons_dataset'
augmented_folder = 'C:/Users/Toshal/Desktop/Project/parkinsons_aug'

# Augment Parkinson's images to approximately 2500 images from 221 images
augment_images(parkinsons_folder, augmented_folder, num_augmented_images=11)
```
